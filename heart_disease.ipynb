# Models

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score,recall_score,precision_score,accuracy_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df_encode.drop('HeartDisease', axis = 1)
y = df_encode['HeartDisease']
```

```python
X_test,X_train,y_test,y_train = train_test_split(X, y , test_size=0.20, random_state=42)
LRModel = LogisticRegression()
LRModel.fit(X_train,y_train)
```

```python
y_pred = LRModel.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
print('f1_score',f1_score(y_test,y_pred))
print('accuracy_score',accuracy_score(y_test,y_pred))
print('recall_score',recall_score(y_test, y_pred))
print('precision_score',precision_score(y_test,y_pred))
print('classification_report',classification_report(y_test, y_pred))
```


***WE can see we got a good recall score up here, which's good thing bcs it's imp to know that accuracy for recall is high***

***Recall -> Measures the proportion of actual positive cases (people who do have heart disease) that were correctly identified by the model.***

***We also got our accuracy and f1_score nice***

```python
X_test,X_train,y_test,y_train = train_test_split(X, y , test_size=0.20, random_state=42)
NVModel = GaussianNB()
NVModel.fit(X_train,y_train)
```

```python
y_pred = NVModel.predict(X_test)
confusion_matrix(y_test,y_pred)
```

```python
print('f1_score : ',f1_score(y_test,y_pred))
print('accuracy_score : ',accuracy_score(y_test,y_pred))
print('recall_score : ',recall_score(y_test, y_pred))
print('precision_score : ',precision_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```
